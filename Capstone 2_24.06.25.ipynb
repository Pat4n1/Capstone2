```python
Q1 = Hallo_Bus_TJ['payAmount'].quantile(0.25)
Q3 = Hallo_Bus_TJ['payAmount'].quantile(0.75)
IQR = Q3 - Q1
```

Identifikasi Outlier

Dari total 37.900 baris data, ditemukan bahwa sebagian besar transaksi memiliki nilai pembayaran Rp 3.500, yang merupakan tarif normal Transjakarta. Namunterdapat sejumlah nilai tidak wajar yang muncul:

Rp 0 → kemungkinan transaksi gagal atau akses gratis (diskon, error sistem, atau kartu error).

Rp 20.000 → jumlah jauh di atas tarif normal, berpotensi salah input atau kondisi khusus.

Outlier diidentifikasi menggunakan metode IQR (Interquartile Range)

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df_clean = Hallo_Bus_TJ[(Hallo_Bus_TJ['payAmount'] >= lower_bound) & (Hallo_Bus_TJ['payAmount'] <= upper_bound)]
```

```python
median_value = Hallo_Bus_TJ['payAmount'].median()
```

```python
Hallo_Bus_TJ['payAmount'] = Hallo_Bus_TJ['payAmount'].apply(
    lambda x: median_value if x < lower_bound or x > upper_bound else x
)
```
